# Covid Data

## Data Collection
Data was downloaded from USA Facts https://usafacts.org/visualizations/coronavirus-covid-19-spread-map/

```python
import pandas as pd
import numpy as np

pd.options.display.max_columns = 320
```

```python
file_path = '../datasets/'
file_covid = '02_covid_confirmed_usafacts.csv'
```

```python
df_original=pd.read_csv(file_path + file_covid)
```

```python
df_original.head(3)
```

```python
# Size of original-cumulative dataframe
df_original.shape
```

```python
date_columns = df_original.drop(columns = ['countyFIPS', 'County Name', 'State', 'stateFIPS']).columns
```

```python
df_location = df_original[['countyFIPS', 'County Name', 'State', 'stateFIPS']]
```

```python
df_dates = df_original[date_columns].copy()
```

```python
covid_cumulative = pd.DataFrame(columns = ['date', 'covid cases','countyFIPS','County Name', 'State', 'stateFIPS'])
```

```python
# This will build a dataframe with the values from the original df in a different format so that it can be used 
#for further analysis
for i in range(3195):
    data = {'date' : df_dates.iloc[i].index,
            'covid cases': df_dates.iloc[i].values,
            'countyFIPS':  np.repeat(df_location.iloc[i][0],315),
            'County Name': np.repeat(df_location.iloc[i][1],315),
            'State':  np.repeat(df_location.iloc[i][2],315),
            'stateFIPS': np.repeat(df_location.iloc[i][3],315),
            }

    temp_df = pd.DataFrame (data, columns = ['date', 'covid cases','countyFIPS','County Name', 'State', 'stateFIPS'])
    covid_cumulative = pd.concat([covid_cumulative, temp_df])
```

```python
covid_cumulative.head()
```

### New Covid Cases
The following code calculates the number of new cases based on the number of cumulative cases from the original dataframe.

```python
new_cases = df_original.copy()
```

```python
def new_cases_calc(df):
    if df[1] > df[0]:
        return df[1]-df[0]
    else:
        return 0
```

```python
for i in range(len(date_columns)-1):
    new_cases[date_columns[i+1]] = df_original[[date_columns[i], date_columns[i+1]]].apply(new_cases_calc, axis=1)
```

```python
new_cases.head()
```

```python
df_dates = new_cases[date_columns].copy()
```

```python
covid_new_cases = pd.DataFrame(columns = ['new covid cases'])
```

```python
for i in range(3195):
    data = {'new covid cases': df_dates.iloc[i].values}
    temp_df = pd.DataFrame (data, columns = ['new covid cases'])
    covid_new_cases = pd.concat([covid_new_cases, temp_df])
```

#### Write cleaned dataframes to csv files

```python
covid_cumulative.to_csv(file_path +'05_cleaned_covid_cumulative.csv', index = True)
covid_new_cases.to_csv(file_path +'06_cleaned_covid_new.csv', index = True)
```
